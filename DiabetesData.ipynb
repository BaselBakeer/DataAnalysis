```python
x=diabetes.drop('Outcome',axis=1)
y=diabetes['Outcome']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.35)

print("Training Data Shape : ", xtrain.shape, ytrain.shape)
print("Test Data Shape : ", xtest.shape, ytest.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(max_depth=20,min_samples_split =20,min_samples_leaf =10)
```

```python
rf.fit(xtrain,ytrain)
```

```python
rf.score(xtrain,ytrain)
```

```python
rf.score(xtest,ytest)
```
