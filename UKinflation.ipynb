```python
monthToNmber={
    "JAN": "-01-01",
    "FEB": '-02-01',
    "MAR": '-03-01',
    "APR": '-04-01',
    "MAY": '-05-01',
    "JUN": '-06-01',
    "JUL": '-07-01',
    "AUG": '-08-01',
    "SEP": '-09-01',
    "OCT": '-10-01',
    "NOV": '-11-01',
    "DEC": '-12-01'

    }
monthlyInflation['Month']=monthlyInflation['Month'].map(monthToNmber)
```

```python
monthlyInflation
```

```python
monthlyInflation['Year'] = monthlyInflation['Year'].astype(str) +monthlyInflation['Month']
```

```python
monthlyInflation
```

```python
monthlyInflation.drop('Month',axis=1,inplace=True)
```

```python
monthlyInflation
```

```python
monthlyInflation['Year'] = pd.to_datetime(monthlyInflation['Year'], yearfirst=True)
```

```python
monthlyInflation.rename(columns={'Year':'Date'},inplace=True)
```
