```python
df=df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Gender=le.fit_transform(df.Gender)
df.Dependents=le.fit_transform(df.Dependents)
df.Education=le.fit_transform(df.Education)
df.Self_Employed=le.fit_transform(df.Self_Employed)
df.Property_Area=le.fit_transform(df.Property_Area)
df.Loan_Status=le.fit_transform(df.Loan_Status)
```

```python
df.Married=le.fit_transform(df.Married)
```

```python
df
```

```python
df.drop('Loan_ID',axis=1,inplace=True)
```

```python
x=df.drop('Loan_Status',axis=1)
y=df['Loan_Status']
from sklearn.model_selection import train_test_split
X_test,X_train,Y_test,Y_train=train_test_split(x,y,test_size=0.2)
```

```python
pipelines = {
    'xgboost': Pipeline([
        ('model', xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42))
    ]),
    'random_forest': Pipeline([
        ('model', RandomForestClassifier(random_state=42))
    ]),
    'decision_tree': Pipeline([
        ('model', DecisionTreeClassifier(random_state=42))
    ]),
    'logistic_regression': Pipeline([
        ('model', LogisticRegression(max_iter=200, random_state=42))
    ]),
    'knn': Pipeline([
        ('model', KNeighborsClassifier())
    ]),
    'gradient_boosting': Pipeline([
        ('model', GradientBoostingClassifier(random_state=42))
    ])
}
```

```python
param_grids = {
    'xgboost': {
        'model__max_depth': [3, 4, 5],
        'model__learning_rate': [0.01, 0.1, 0.2],
        'model__n_estimators': [50, 100]
    },
    'random_forest': {
        'model__n_estimators': [50, 100],
        'model__max_depth': [3, 4, 5],
        'model__min_samples_split': [2, 5]
    },
    'decision_tree': {
        'model__max_depth': [3, 4, 5, None],
        'model__min_samples_split': [2, 5, 10]
    },
    'logistic_regression': {
        'model__C': [0.01, 0.1, 1, 10],
        'model__solver': ['lbfgs', 'liblinear'],
        'model__penalty': ['l2']
    },
    'knn': {
        'model__n_neighbors': [3, 5, 7, 9],
        'model__weights': ['uniform', 'distance'],
        'model__metric': ['euclidean', 'manhattan']
    },
    'gradient_boosting': {
        'model__n_estimators': [50, 100],
        'model__learning_rate': [0.01, 0.1, 0.2],
        'model__max_depth': [3, 4, 5],
        'model__min_samples_split': [2, 5]
    }
}
```

```python
best_models = {}
for model_name, pipeline in pipelines.items():
    grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grids[model_name], 
                               scoring='accuracy', cv=3, verbose=1, n_jobs=-1)
    grid_search.fit(X_train, Y_train)
    best_models[model_name] = grid_search.best_estimator_
    print(f"Best Parameters for {model_name}: {grid_search.best_params_}")
```

```python
for model_name, best_model in best_models.items():
    # Calculate training accuracy
    training_accuracy = best_model.score(X_train, Y_train)  # Training accuracy
    # Calculate test accuracy
    y_pred = best_model.predict(X_test)  # Predictions on the test set
    accuracy = accuracy_score(Y_test, y_pred)  # Test accuracy
    print(f"Training Accuracy of {model_name}: {training_accuracy * 100:.2f}%")
    print(f"Test Accuracy of {model_name}: {accuracy * 100:.2f}%")
```
